```python
#Libraries
import plotly.graph_objects as go
```

```python
#initialization

#define x1, x2, y For the first time.
p1 = [0,0]
p2 = [0,1000]
p3 = [1000,1000]
p4 = [1000,0]
fig = go.Figure(go.Scatter( fill="toself"))

#fix the aspect ratio
fig.update_xaxes(scaleanchor="y", scaleratio=1)
```

```python
#Looping
for i in range(0,30):

    #create a square  1000 x 1000
    fig.add_trace(go.Scatter(x=[p1[0],p2[0],p3[0],p4[0],p1[0]], y=[p1[1],p2[1],p3[1],p4[1],p1[1]], fill="toself", fillcolor="blue"))

    #updating the square to be the up two thirds
    new_y= p1[1] + (abs(p2[1]-p1[1])/3)
    Sqp1 = [p1[0],new_y]
    Sqp2 = p2
    Sqp3 = p3
    Sqp4 = [p4[0],new_y]

    #plotting the new square in the same figure 

    fig.add_trace(go.Scatter(x=[Sqp1[0],Sqp2[0],Sqp3[0],Sqp4[0]], y=[Sqp1[1],Sqp2[1],Sqp3[1],Sqp4[1]], fill="toself", fillcolor="blue"))  #2/3
    fig.add_trace(go.Scatter(x=[p1[0],Sqp1[0],Sqp4[0],p4[0]], y=[p1[1],Sqp1[1],Sqp4[1],p4[1]], fill="toself", fillcolor="red"))           #1/3

    #adding two triangles to the figure in the corners

      #triangle 1
    Trip1 = p1
    Trip2 = Sqp1
    Trip3 = [Sqp1[0] + abs(Sqp4[0]-Sqp1[0])/3,Sqp1[1]]
    fig.add_trace(go.Scatter(x=[Trip1[0],Trip2[0],Trip3[0],Trip1[0]], y=[Trip1[1],Trip2[1],Trip3[1],Trip1[1]], fill="toself", fillcolor="blue"))


      #triangle 2
    Tri2p1 = p4
    Tri2p2 = Sqp4
    Tri2p3 = [Sqp1[0]+ 2*abs(Sqp4[0]-Sqp1[0])/3,Sqp4[1]]
    fig.add_trace(go.Scatter(x=[Tri2p1[0],Tri2p2[0],Tri2p3[0],Tri2p1[0]], y=[Tri2p1[1],Tri2p2[1],Tri2p3[1],Tri2p1[1]], fill="toself", fillcolor="blue"))


    #update  new square coordinates
    new_y_sq = abs(Trip3[0] - Tri2p3[0])
    print(new_y_sq)                       #Printing the height of the new trapezoid (- Area) = height of the triangles (+area) : n--> oo : h=0 (nothing added or substracted --> Converges)
    p1 = Trip3
    p2 = [Trip3[0],Trip3[1]+new_y_sq]
    p3 = [Tri2p3[0],Tri2p3[1]+new_y_sq]
    p4 = Tri2p3

    fig.show()

    if(new_y_sq<10**-4):
      break
```
